```python
def extract_frames(video_path, num_frames=10):
    cap = cv2.VideoCapture(video_path)
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))

    frame_interval = max(total_frames // num_frames, 1)
    frames = []
    frame_index = 0
    frames_extracted = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        if frame_index % frame_interval == 0:
            frames.append(frame)
            frames_extracted += 1

        frame_index += 1
        if frames_extracted >= num_frames:
            break

    cap.release()
    return frames
```

```python
import os
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from PIL import Image
from torch.optim.lr_scheduler import ReduceLROnPlateau
import warnings
warnings.filterwarnings("ignore")

# ======================== CONFIGURACION ========================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
frames_per_video = 10
batch_size = 8
img_size = (120, 160)

# ======================== TRANSFORMACIONES ========================
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# ======================== DATASET PERSONALIZADO ========================
class VideoDataset(Dataset):
    def __init__(self, root_dir, transform=None, frames_per_video=10):
        self.root_dir = root_dir
        self.transform = transform
        self.frames_per_video = frames_per_video
        self.video_paths = []
        self.labels = []
        self.classes = sorted(os.listdir(root_dir))
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}

        for class_name in self.classes:
            class_path = os.path.join(root_dir, class_name)
            for video_folder in os.listdir(class_path):
                video_path = os.path.join(class_path, video_folder)
                frames = sorted([os.path.join(video_path, f) for f in os.listdir(video_path) if f.endswith('.jpg')])
                if len(frames) >= frames_per_video:
                    self.video_paths.append(frames[:frames_per_video])
                    self.labels.append(self.class_to_idx[class_name])

    def __len__(self):
        return len(self.video_paths)

    def __getitem__(self, idx):
        frame_paths = self.video_paths[idx]
        frames = [self.transform(Image.open(p).convert('RGB')) for p in frame_paths]
        video_tensor = torch.stack(frames)  # (T, C, H, W)
        label = self.labels[idx]
        return video_tensor, label

# ======================== LSTM MANUAL ========================
class net_LSTM_manual(nn.Module):
    def __init__(self, x_size, a_size, y_size):
        super(net_LSTM_manual, self).__init__()
        self.a_size = a_size
        self.x_size = x_size

        self.W_f = nn.Linear(x_size + a_size, a_size)
        self.W_i = nn.Linear(x_size + a_size, a_size)
        self.W_C = nn.Linear(x_size + a_size, a_size)
        self.W_o = nn.Linear(x_size + a_size, a_size)

        self.tr_lin_y = nn.Linear(a_size, y_size)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, x, hidden_state):
        h_prev, C_prev = hidden_state
        x = x.view(1, -1)
        combined = torch.cat((x, h_prev), dim=1)

        f_t = torch.sigmoid(self.W_f(combined))
        i_t = torch.sigmoid(self.W_i(combined))
        C_tilde = torch.tanh(self.W_C(combined))
        C_t = f_t * C_prev + i_t * C_tilde
        o_t = torch.sigmoid(self.W_o(combined))
        h_t = o_t * torch.tanh(C_t)

        output = self.tr_lin_y(h_t)
        output = self.softmax(output)

        return output, (h_t, C_t)

    def init_hidden(self):
        return (torch.zeros(1, self.a_size).to(device), torch.zeros(1, self.a_size).to(device))

# ======================== CNN + LSTM MANUAL ========================
class CNN_LSTM(nn.Module):
    def __init__(self, hidden_dim, output_dim):
        super(CNN_LSTM, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1), nn.ReLU(),nn.MaxPool2d(2),
            nn.Conv2d(16, 32, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.flatten_dim = 64 * (img_size[0]//8) * (img_size[1]//8)
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.lstm_cell = net_LSTM_manual(self.flatten_dim, hidden_dim, output_dim)

    def forward(self, x):
        B, T, C, H, W = x.size()
        x = x.view(B * T, C, H, W)
        x = self.cnn(x)
        x = x.view(B, T, -1)

        outputs = []
        for b in range(B):
            h_t, c_t = self.lstm_cell.init_hidden()
            for t in range(T):
                out, (h_t, c_t) = self.lstm_cell(x[b, t], (h_t, c_t))
            outputs.append(out)
        return torch.stack(outputs).squeeze(1)

# ======================== CARGA DE DATOS ========================
train_dataset = VideoDataset('/content/dataset/train', transform=transform)
valid_dataset = VideoDataset('/content/dataset/valid', transform=transform)
test_dataset  = VideoDataset('/content/dataset/test', transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

# ======================== ENTRENAMIENTO ========================
def train(model, train_loader, valid_loader, num_epochs=50):
    model.to(device)
    criterion = nn.NLLLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)

    best_val_loss = float('inf')
    best_model_state = None

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for videos, labels in train_loader:
            videos, labels = videos.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(videos)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for videos, labels in valid_loader:
                videos, labels = videos.to(device), labels.to(device)
                outputs = model(videos)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        scheduler.step(val_loss)
        print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()

    if best_model_state:
        torch.save(best_model_state, "best_cnn_lstm_model.pt")
        print("\nMejor modelo guardado como 'best_cnn_lstm_model.pt'")

# ======================== ENTRENAR ========================
output_dim = len(train_dataset.classes)
model = CNN_LSTM(hidden_dim=256, output_dim=output_dim)
# Solo entrenamos si el modelo no ha sido entrenado o si quieres re-entrenar
train(model, train_loader, valid_loader) # Uncomment this line to train the model

# ======================== EVALUACION Y MATRICES DE CONFUSION ========================
model.load_state_dict(torch.load("best_cnn_lstm_model.pt"))
model.to(device)
model.eval()

def evaluate_and_plot_confusion_matrix(model, data_loader, dataset_name, classes):
    predicted_all = []
    actual_all = []

    with torch.no_grad():
        for videos, labels in data_loader:
            outputs = model(videos.to(device))
            preds = torch.argmax(outputs, dim=1).cpu()
            predicted_all.extend(preds.tolist())
            actual_all.extend(labels.tolist())

    print(f"\nReporte de Clasificación ({dataset_name}):")
    print(classification_report(actual_all, predicted_all, target_names=classes))

    conf_mat = confusion_matrix(actual_all, predicted_all)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicción')
    plt.ylabel('Valor Real')
    plt.title(f'Matriz de Confusión ({dataset_name})')
    plt.tight_layout()
    plt.show()

# Evaluar y mostrar matriz de confusión para cada dataset
evaluate_and_plot_confusion_matrix(model, train_loader, "Train", train_dataset.classes)
evaluate_and_plot_confusion_matrix(model, valid_loader, "Validación", train_dataset.classes)
evaluate_and_plot_confusion_matrix(model, test_loader, "Test", train_dataset.classes)
```

```python
import os
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from PIL import Image
from torch.optim.lr_scheduler import ReduceLROnPlateau
import warnings
warnings.filterwarnings("ignore")

# ======================== CONFIGURACION ========================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
frames_per_video = 10
batch_size = 8  # Keeping batch size for now, can be tuned
img_size = (120, 160)

# ======================== TRANSFORMACIONES ========================
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# ======================== DATASET PERSONALIZADO ========================
class VideoDataset(Dataset):
    def __init__(self, root_dir, transform=None, frames_per_video=10):
        self.root_dir = root_dir
        self.transform = transform
        self.frames_per_video = frames_per_video
        self.video_paths = []
        self.labels = []
        self.classes = sorted(os.listdir(root_dir))
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}

        for class_name in self.classes:
            class_path = os.path.join(root_dir, class_name)
            for video_folder in os.listdir(class_path):
                video_path = os.path.join(class_path, video_folder)
                frames = sorted([os.path.join(video_path, f) for f in os.listdir(video_path) if f.endswith('.jpg')])
                if len(frames) >= frames_per_video:
                    self.video_paths.append(frames[:frames_per_video])
                    self.labels.append(self.class_to_idx[class_name])

    def __len__(self):
        return len(self.video_paths)

    def __getitem__(self, idx):
        frame_paths = self.video_paths[idx]
        frames = [self.transform(Image.open(p).convert('RGB')) for p in frame_paths]
        video_tensor = torch.stack(frames)  # (T, C, H, W)
        label = self.labels[idx]
        return video_tensor, label

# ======================== LSTM MANUAL ========================
class net_LSTM_manual(nn.Module):
    def __init__(self, x_size, a_size, y_size):
        super(net_LSTM_manual, self).__init__()
        self.a_size = a_size
        self.x_size = x_size

        self.W_f = nn.Linear(x_size + a_size, a_size)
        self.W_i = nn.Linear(x_size + a_size, a_size)
        self.W_C = nn.Linear(x_size + a_size, a_size)
        self.W_o = nn.Linear(x_size + a_size, a_size)

        self.tr_lin_y = nn.Linear(a_size, y_size)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, x, hidden_state):
        h_prev, C_prev = hidden_state
        x = x.view(1, -1)
        combined = torch.cat((x, h_prev), dim=1)

        f_t = torch.sigmoid(self.W_f(combined))
        i_t = torch.sigmoid(self.W_i(combined))
        C_tilde = torch.tanh(self.W_C(combined))
        C_t = f_t * C_prev + i_t * C_tilde
        o_t = torch.sigmoid(self.W_o(combined))
        h_t = o_t * torch.tanh(C_t)

        output = self.tr_lin_y(h_t)
        output = self.softmax(output)

        return output, (h_t, C_t)

    def init_hidden(self):
        return (torch.zeros(1, self.a_size).to(device), torch.zeros(1, self.a_size).to(device))

# ======================== CNN + LSTM MANUAL (with Dropout) ========================
class CNN_LSTM(nn.Module):
    def __init__(self, hidden_dim, output_dim, dropout_prob=0.5):
        super(CNN_LSTM, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2), nn.Dropout(dropout_prob),
            nn.Conv2d(16, 32, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2), nn.Dropout(dropout_prob),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2), nn.Dropout(dropout_prob),
        )
        self.flatten_dim = 64 * (img_size[0]//8) * (img_size[1]//8)
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.lstm_cell = net_LSTM_manual(self.flatten_dim, hidden_dim, output_dim)
        self.dropout_lstm_output = nn.Dropout(dropout_prob) # Dropout after LSTM output

    def forward(self, x):
        B, T, C, H, W = x.size()
        x = x.view(B * T, C, H, W)
        x = self.cnn(x)
        x = x.view(B, T, -1)

        outputs = []
        for b in range(B):
            h_t, c_t = self.lstm_cell.init_hidden()
            for t in range(T):
                out, (h_t, c_t) = self.lstm_cell(x[b, t], (h_t, c_t))
            outputs.append(out)

        final_output = torch.stack(outputs).squeeze(1)
        final_output = self.dropout_lstm_output(final_output) # Apply dropout to the final output

        return final_output

# ======================== CARGA DE DATOS ========================
train_dataset = VideoDataset('/content/dataset/train', transform=transform)
valid_dataset = VideoDataset('/content/dataset/valid', transform=transform)
test_dataset  = VideoDataset('/content/dataset/test', transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

# ======================== ENTRENAMIENTO ========================
def train(model, train_loader, valid_loader, num_epochs=100):
    model.to(device)
    criterion = nn.NLLLoss()
    # Switched to AdamW and slightly adjusted initial learning rate and weight_decay
    optimizer = torch.optim.AdamW(model.parameters(), lr=0.0005, weight_decay=1e-5)
    # Adjusted scheduler parameters: factor and patience
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.3, patience=7, verbose=True)
    gradient_clip_val = 1.0 # Value for gradient clipping

    best_val_loss = float('inf')
    best_model_state = None

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for videos, labels in train_loader:
            videos, labels = videos.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(videos)
            loss = criterion(outputs, labels)
            loss.backward()

            # Apply gradient clipping
            torch.nn.utils.clip_grad_norm_(model.parameters(), gradient_clip_val)

            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for videos, labels in valid_loader:
                videos, labels = videos.to(device), labels.to(device)
                outputs = model(videos)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        scheduler.step(val_loss)
        print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()

    if best_model_state:
        torch.save(best_model_state, "best_cnn_lstm_model.pt")
        print("\nMejor modelo guardado como 'best_cnn_lstm_model.pt'")

# ======================== ENTRENAR ========================
output_dim = len(train_dataset.classes)
model = CNN_LSTM(hidden_dim=256, output_dim=output_dim, dropout_prob=0.5)
train(model, train_loader, valid_loader, num_epochs=100)

# ======================== EVALUACION Y MATRICES DE CONFUSION ========================
model.load_state_dict(torch.load("best_cnn_lstm_model.pt"))
model.to(device)
model.eval()

def evaluate_and_plot_confusion_matrix(model, data_loader, dataset_name, classes):
    predicted_all = []
    actual_all = []

    with torch.no_grad():
        for videos, labels in data_loader:
            outputs = model(videos.to(device))
            preds = torch.argmax(outputs, dim=1).cpu()
            predicted_all.extend(preds.tolist())
            actual_all.extend(labels.tolist())

    print(f"\nReporte de Clasificación ({dataset_name}):")
    print(classification_report(actual_all, predicted_all, target_names=classes))

    conf_mat = confusion_matrix(actual_all, predicted_all)
    plt.figure(figsize=(8, 6))
    sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.xlabel('Predicción')
    plt.ylabel('Valor Real')
    plt.title(f'Matriz de Confusión ({dataset_name})')
    plt.tight_layout()
    plt.show()

# Evaluar y mostrar matriz de confusión para cada dataset
evaluate_and_plot_confusion_matrix(model, train_loader, "Train", train_dataset.classes)
evaluate_and_plot_confusion_matrix(model, valid_loader, "Validación", train_dataset.classes)
evaluate_and_plot_confusion_matrix(model, test_loader, "Test", train_dataset.classes)
```

```python


import os
import cv2
import torch
import torch.nn as nn
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from torchvision import transforms
from torch.utils.data import Dataset, DataLoader
from sklearn.metrics import classification_report, confusion_matrix
from PIL import Image
from torch.optim.lr_scheduler import ReduceLROnPlateau
import warnings
warnings.filterwarnings("ignore")

# ======================== CONFIGURACION ========================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
frames_per_video = 10
batch_size = 8
img_size = (120, 160)

# ======================== TRANSFORMACIONES ========================
transform = transforms.Compose([
    transforms.Resize(img_size),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# ======================== DATASET PERSONALIZADO ========================
class VideoDataset(Dataset):
    def __init__(self, root_dir, transform=None, frames_per_video=10):
        self.root_dir = root_dir
        self.transform = transform
        self.frames_per_video = frames_per_video
        self.video_paths = []
        self.labels = []
        self.classes = sorted(os.listdir(root_dir))
        self.class_to_idx = {cls_name: i for i, cls_name in enumerate(self.classes)}

        for class_name in self.classes:
            class_path = os.path.join(root_dir, class_name)
            for video_folder in os.listdir(class_path):
                video_path = os.path.join(class_path, video_folder)
                frames = sorted([os.path.join(video_path, f) for f in os.listdir(video_path) if f.endswith('.jpg')])
                if len(frames) >= frames_per_video:
                    self.video_paths.append(frames[:frames_per_video])
                    self.labels.append(self.class_to_idx[class_name])

    def __len__(self):
        return len(self.video_paths)

    def __getitem__(self, idx):
        frame_paths = self.video_paths[idx]
        frames = [self.transform(Image.open(p).convert('RGB')) for p in frame_paths]
        video_tensor = torch.stack(frames)  # (T, C, H, W)
        label = self.labels[idx]
        return video_tensor, label

# ======================== LSTM MANUAL ========================
class net_LSTM_manual(nn.Module):
    def __init__(self, x_size, a_size, y_size):
        super(net_LSTM_manual, self).__init__()
        self.a_size = a_size
        self.x_size = x_size

        self.W_f = nn.Linear(x_size + a_size, a_size)
        self.W_i = nn.Linear(x_size + a_size, a_size)
        self.W_C = nn.Linear(x_size + a_size, a_size)
        self.W_o = nn.Linear(x_size + a_size, a_size)

        self.tr_lin_y = nn.Linear(a_size, y_size)
        self.softmax = nn.LogSoftmax(dim=1)

    def forward(self, x, hidden_state):
        h_prev, C_prev = hidden_state
        x = x.view(1, -1)
        combined = torch.cat((x, h_prev), dim=1)

        f_t = torch.sigmoid(self.W_f(combined))
        i_t = torch.sigmoid(self.W_i(combined))
        C_tilde = torch.tanh(self.W_C(combined))
        C_t = f_t * C_prev + i_t * C_tilde
        o_t = torch.sigmoid(self.W_o(combined))
        h_t = o_t * torch.tanh(C_t)

        output = self.tr_lin_y(h_t)
        output = self.softmax(output)

        return output, (h_t, C_t)

    def init_hidden(self):
        return (torch.zeros(1, self.a_size).to(device), torch.zeros(1, self.a_size).to(device))

# ======================== CNN + LSTM MANUAL ========================
class CNN_LSTM(nn.Module):
    def __init__(self, hidden_dim, output_dim):
        super(CNN_LSTM, self).__init__()
        self.cnn = nn.Sequential(
            nn.Conv2d(3, 16, kernel_size=3, padding=1), nn.ReLU(),nn.MaxPool2d(2),
            nn.Conv2d(16, 32, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, kernel_size=3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.flatten_dim = 64 * (img_size[0]//8) * (img_size[1]//8)
        self.hidden_dim = hidden_dim
        self.output_dim = output_dim
        self.lstm_cell = net_LSTM_manual(self.flatten_dim, hidden_dim, output_dim)

    def forward(self, x):
        B, T, C, H, W = x.size()
        x = x.view(B * T, C, H, W)
        x = self.cnn(x)
        x = x.view(B, T, -1)

        outputs = []
        for b in range(B):
            h_t, c_t = self.lstm_cell.init_hidden()
            for t in range(T):
                out, (h_t, c_t) = self.lstm_cell(x[b, t], (h_t, c_t))
            outputs.append(out)
        return torch.stack(outputs).squeeze(1)

# ======================== CARGA DE DATOS ========================
train_dataset = VideoDataset('/content/dataset/train', transform=transform)
valid_dataset = VideoDataset('/content/dataset/valid', transform=transform)
test_dataset  = VideoDataset('/content/dataset/test', transform=transform)

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
valid_loader = DataLoader(valid_dataset, batch_size=batch_size, shuffle=False)
test_loader  = DataLoader(test_dataset,  batch_size=batch_size, shuffle=False)

# ======================== ENTRENAMIENTO ========================
def train(model, train_loader, valid_loader, num_epochs=50):
    model.to(device)
    criterion = nn.NLLLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=0.001, weight_decay=1e-4)
    scheduler = ReduceLROnPlateau(optimizer, mode='min', factor=0.5, patience=5, verbose=True)

    best_val_loss = float('inf')
    best_model_state = None

    for epoch in range(num_epochs):
        model.train()
        train_loss = 0
        for videos, labels in train_loader:
            videos, labels = videos.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(videos)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            train_loss += loss.item()

        model.eval()
        val_loss = 0
        with torch.no_grad():
            for videos, labels in valid_loader:
                videos, labels = videos.to(device), labels.to(device)
                outputs = model(videos)
                loss = criterion(outputs, labels)
                val_loss += loss.item()

        scheduler.step(val_loss)
        print(f"Epoch {epoch+1}/{num_epochs} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f}")

        if val_loss < best_val_loss:
            best_val_loss = val_loss
            best_model_state = model.state_dict()

    if best_model_state:
        torch.save(best_model_state, "best_cnn_lstm_model.pt")
        print("\nMejor modelo guardado como 'best_cnn_lstm_model.pt'")

# ======================== ENTRENAR ========================
output_dim = len(train_dataset.classes)
model = CNN_LSTM(hidden_dim=256, output_dim=output_dim)
train(model, train_loader, valid_loader)

# ======================== EVALUACION ========================
model.load_state_dict(torch.load("best_cnn_lstm_model.pt"))
model.to(device)
model.eval()
predicted_all = []
actual_all = []

with torch.no_grad():
    for videos, labels in test_loader:
        outputs = model(videos.to(device))
        preds = torch.argmax(outputs, dim=1).cpu()
        predicted_all.extend(preds.tolist())
        actual_all.extend(labels.tolist())

print("\nReporte de Clasificación:")
print(classification_report(actual_all, predicted_all, target_names=train_dataset.classes))

conf_mat = confusion_matrix(actual_all, predicted_all)
plt.figure(figsize=(8, 6))
sns.heatmap(conf_mat, annot=True, fmt='d', cmap='Blues', xticklabels=train_dataset.classes, yticklabels=train_dataset.classes)
plt.xlabel('Predicción')
plt.ylabel('Valor Real')
plt.title('Matriz de Confusión')
plt.tight_layout()
plt.show()
```
